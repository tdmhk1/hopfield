```python
epsilon = 0.1
outputs = tspnet.outputs.copy().astype(np.int64)
for i in range(N):
    for j in range(N):
        x = outputs[i][j]
        if x > 1 - epsilon:
            outputs[i][j] = 1
        elif x < epsilon:
            outputs[i][j] = 0
        else:
            raise NameError('cannot decide')
```
