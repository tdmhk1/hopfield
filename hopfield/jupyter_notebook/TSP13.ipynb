```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.load("points2.npy")
```

```python
plt.plot(points[:, 0], points[:, 1], "o", color="r")
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
plt.axes().set_aspect('equal')
plt.xlabel("N = 10", fontsize=20)
plt.savefig("points.png")
plt.show()
```

```python
route = [0, 5, 8, 9, 6, 4, 7, 1, 2, 3, 0]
```

```python
import sys
sys.path.append("./TSP/")
```

```python
import TSP
```

```python
dists = TSP.calc_dists(points)
```

```python
plt.plot(points[:, 0], points[:, 1], "o", color="r")
plt.plot([points[i, 0] for i in route], [points[i, 1] for i in route], color="b")
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
plt.axes().set_aspect('equal')
plt.xlabel("total = "+ str(TSP.calc_route_dist(dists, route)), fontsize=20)
plt.savefig("minimum_route2.png")
plt.show()
```

```python
results = []
for i in range(1000):
    tspnet = TSP.TSPNetwork(size=10, A=500, B=500, C=200, D=500, n=15, u_0=10, dists=dists)
    tspnet.update()
    try:
        route = tspnet.decide_route()
        route.append(route[0])
        results.append(TSP.calc_route_dist(dists, route))
    except:
        results.append(False)
    print("epoch" + str(i))
```

```python
results2 = []
for i in results:
    if i:
        results2.append(i)
```
